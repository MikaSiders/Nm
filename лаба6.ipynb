```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# --- 1. Вхідні дані (Варіант 30) ---
x = np.array([-3, -2, -1, 0], dtype=float)
y = np.array([4, 4, 0, -2], dtype=float)

# Точки, в яких треба знайти значення (із завдання)
x_targets = [-4, -3.5, 0.5, 1]

# --- 2. Побудова багаточлена Лагранжа ---
poly = lagrange(x, y)

print("Інтерполяційний багаточлен Лагранжа:\n")
print(poly) # Виводить коефіцієнти полінома

print("\n--- Обчислені значення у заданих точках ---")
for xt in x_targets:
    yt = poly(xt)
    print(f"x = {xt:<5} -> y = {yt:.4f}")

# --- 3. Побудова графіка ---
# Створюємо масив точок для плавної лінії графіка
x_plot = np.linspace(min(x) - 1.5, max(x) + 1.5, 100)
y_plot = poly(x_plot)

plt.figure(figsize=(10, 6))

# Графік полінома
plt.plot(x_plot, y_plot, label='Багаточлен Лагранжа', color='blue')

# Вхідні вузли інтерполяції
plt.scatter(x, y, color='red', s=100, zorder=5, label='Вузли інтерполяції')

# Шукані точки
y_targets = poly(x_targets)
plt.scatter(x_targets, y_targets, color='green', marker='x', s=100, zorder=5, label='Шукані точки')

plt.title('Інтерполяція Багаточлен Лагранжа')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
