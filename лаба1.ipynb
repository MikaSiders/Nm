```python
import numpy as np

# --- ЗАВДАННЯ 1 ---
print("--- ЗАВДАННЯ 1 ---")
# Дані
val1_approx = 7.28
val1_exact = np.sqrt(53)
val2_approx = 0.823
val2_exact = 14/17

# Похибки
abs_err1 = abs(val1_approx - val1_exact)
rel_err1 = abs_err1 / val1_exact

abs_err2 = abs(val2_approx - val2_exact)
rel_err2 = abs_err2 / val2_exact

print(f"1) sqrt(53) ≈ {val1_approx}")
print(f"   Абс. похибка: {abs_err1:.6f}")
print(f"   Відн. похибка: {rel_err1:.7f} ({rel_err1*100:.4f}%)")

print(f"2) 14/17 ≈ {val2_approx}")
print(f"   Абс. похибка: {abs_err2:.6f}")
print(f"   Відн. похибка: {rel_err2:.7f} ({rel_err2*100:.4f}%)")

if rel_err1 < rel_err2:
    print("Висновок: Перша рівність точніша.")
else:
    print("Висновок: Друга рівність точніша.")

# --- ЗАВДАННЯ 2 ---
print("\n--- ЗАВДАННЯ 2 ---")
# a) Вузьке розуміння
num_a = 0.66835
delta_a = 0.00115
print(f"a) Число: {num_a} (+- {delta_a})")
# Перевірка округлення до 2 знаків (0.67)
rounded_a = round(num_a, 2)
round_err_a = abs(rounded_a - num_a)
total_err_a = delta_a + round_err_a
limit_a = 0.5 * (10**-2) # половина одиниці сотих
print(f"   Округлення до: {rounded_a}")
print(f"   Загальна похибка: {total_err_a:.5f}")
print(f"   Межа (вузьке): {limit_a}")
print(f"   Умова виконується: {total_err_a <= limit_a}")

# б) Широке розуміння
num_b = 23.3748
rel_delta_b = 0.0027 # 0.27%
abs_delta_b = num_b * rel_delta_b
print(f"б) Число: {num_b}, похибка: {abs_delta_b:.5f}")
# Перевірка округлення до 1 знаку (23.4)
rounded_b = round(num_b, 1)
round_err_b = abs(rounded_b - num_b)
total_err_b = abs_delta_b + round_err_b
limit_b = 1 * (10**-1) # одиниця десятих
print(f"   Округлення до: {rounded_b}")
print(f"   Загальна похибка: {total_err_b:.5f}")
print(f"   Межа (широке): {limit_b}")
print(f"   Умова виконується: {total_err_b <= limit_b}")

# --- ЗАВДАННЯ 3 ---
print("\n--- ЗАВДАННЯ 3 ---")
# а) 43.813 (вузьке)
val_3a = 43.813
abs_err_3a = 0.5 * (10**-3)
rel_err_3a = abs_err_3a / val_3a
print(f"a) Число {val_3a} (вузьке)")
print(f"   Абс. похибка: {abs_err_3a}")
print(f"   Відн. похибка: {rel_err_3a:.7f} ({rel_err_3a*100:.5f}%)")

# б) 0.64 (широке)
val_3b = 0.64
abs_err_3b = 1 * (10**-2)
rel_err_3b = abs_err_3b / val_3b
print(f"б) Число {val_3b} (широке)")
print(f"   Абс. похибка: {abs_err_3b}")
print(f"   Відн. похибка: {rel_err_3b:.6f} ({rel_err_3b*100:.4f}%)")
```
