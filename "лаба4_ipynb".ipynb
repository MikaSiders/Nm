```python
import numpy as np

# --- Вхідні дані (Варіант 30) ---
A = np.matrix([
    [4, 1, 4],
    [1, 1, 2],
    [2, 1, 2]
], dtype=float)

B = np.matrix([
    [-2],
    [-1],
    [0]
], dtype=float)

print("Матриця A:\n", A)
print("Вектор B:\n", B)
print("-" * 30)

# --- 1. Метод Крамера ---
def solve_cramer(a, b):
    print("\n--- Метод Крамера ---")
    det_a = np.linalg.det(a)
    print(f"Головний визначник (det A) = {det_a:.4f}")

    if np.isclose(det_a, 0):
        print("Визначник дорівнює нулю, метод Крамера не застосовується.")
        return None

    # Створення допоміжних матриць
    a_x = a.copy()
    a_x[:, 0] = b  # Заміна 1-го стовпця

    a_y = a.copy()
    a_y[:, 1] = b  # Заміна 2-го стовпця

    a_z = a.copy()
    a_z[:, 2] = b  # Заміна 3-го стовпця

    det_x = np.linalg.det(a_x)
    det_y = np.linalg.det(a_y)
    det_z = np.linalg.det(a_z)

    print(f"det X = {det_x:.4f}")
    print(f"det Y = {det_y:.4f}")
    print(f"det Z = {det_z:.4f}")

    x = det_x / det_a
    y = det_y / det_a
    z = det_z / det_a

    print(f"Результат: x = {x:.4f}, y = {y:.4f}, z = {z:.4f}")
    return x, y, z

# --- 2. Матричний метод (Обернена матриця) ---
def solve_matrix_method(a, b):
    print("\n--- Матричний метод ---")
    try:
        a_inv = np.linalg.inv(a)
        print("Обернена матриця A^(-1):\n", a_inv)

        X = a_inv.dot(b)
        print(f"Результат (X = A^(-1) * B):\n{X}")
        return X
    except np.linalg.LinAlgError:
        print("Матриця вироджена, оберненої матриці не існує.")
        return None

# --- 3. Метод Гауса ---
def solve_gauss(a_in, b_in):
    print("\n--- Метод Гауса ---")
    # Копіюємо, щоб не змінювати глобальні змінні
    a = np.copy(a_in)
    b = np.copy(b_in)
    n = len(b)
    x = np.zeros(n)

    # Прямий хід
    for k in range(n - 1):
        for i in range(k + 1, n):
            if a[k, k] == 0:
                continue # Спрощена обробка ділення на нуль
            factor = a[i, k] / a[k, k]
            for j in range(k, n):
                a[i, j] = a[i, j] - factor * a[k, j]
            b[i] = b[i] - factor * b[k]

    # Зворотний хід
    x[n - 1] = b[n - 1] / a[n - 1, n - 1]
    for i in range(n - 2, -1, -1):
        sum_ax = 0
        for j in range(i + 1, n):
            sum_ax += a[i, j] * x[j]
        x[i] = (b[i] - sum_ax) / a[i, i]

    print(f"Результат: {x}")
    return x

# --- Виклик функцій ---
solve_cramer(A, B)
solve_matrix_method(A, B)
solve_gauss(A, B)

# Перевірка вбудованим засобом
print("\n--- Перевірка (np.linalg.solve) ---")
print(np.linalg.solve(A, B))
```
